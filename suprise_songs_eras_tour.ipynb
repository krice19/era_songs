```python
from config import CLIENT_ID, CLIENT_SECRET, REDIRECT_URI
import json
import requests
import pandas as pd
```

```python
BASE_URL = 'https://api.spotify.com/v1/tracks/'
AUTH_URL = 'https://accounts.spotify.com/api/token'
```

```python
df = pd.read_csv('surprise_songs.csv')
```

```python
df = df.rename(columns={'Suprise Song 2 Code': 'Suprise Song 2 URI'})
```

```python
surprise1 = df['Suprise Song 1 URI'].values.tolist()
surprise2 = df['Suprise Song 2 URI'].values.tolist()
```

```python
# POST

auth_response = requests.post(AUTH_URL, {
    'grant_type': 'client_credentials',
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
})

# convert the response to JSON
auth_response_data = auth_response.json()

# save the access token
access_token = auth_response_data['access_token']
```

```python
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}
```

```python
surprise1_info = []
for i in suprise1:
    res = requests.get(BASE_URL+i, headers=headers)
    
    surprise1_info.append(res.json())

```

```python
surprise1_pop = []
```

```python
for i in range (0,len(surprise1_info)):
    surprise1_pop.append(surprise1_info[i]['popularity'])
```

```python
surprise1_pop
```

```python
df["Surprise Night 1 Popularity"] = surprise1_pop
```

```python
surprise2_info = []
surprise2_pop = []

for i in suprise2:
    res = requests.get(BASE_URL+i, headers=headers)
    
    surprise2_info.append(res.json())
    
for i in range (0,len(surprise2_info)):
    surprise2_pop.append(surprise2_info[i]['popularity'])

```

```python
df["Surprise Night 2 Popularity"] = surprise2_pop
```
